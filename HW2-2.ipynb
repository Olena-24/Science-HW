```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

cnt = df.isnull().sum() / len(df) 
cnt.round(2)
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця Язык.программирования
na_columns = df.isna().any()
columns_to_keep = ['Мова програмування']
na_columns[columns_to_keep] = False
df.drop(na_columns.index[na_columns], axis=1, inplace=True)
df
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець 
# "Мова.програмування"

cnt = df.isnull().sum() / len(df)
cnt.round(2)
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df = df.dropna()
df.head()
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

df.groupby("Посада")
df
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", 
# виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у 
# стовпчику "Зарплата.в.місяць"
new_df = df
new_df = df.groupby("Посада")['Зарплата на місяць'].agg(['min', 'max'])
new_df
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. 
# Використовуйте її для методу apply та створіть новий стовпчик "avg"


def fill_avg_salary(salary):
    return salary.mean()

new_df['average salary'] = new_df.apply(fill_avg_salary, axis=1)
new_df
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
new_df["average salary"].describe()
```

```python
# Збережіть отриману таблицю в CSV файл
new_df.to_csv('2017_jun_final_new.csv', encoding='utf8')
```

```python
#График 2
sns.set_style('darkgrid')
df = pd.DataFrame(new_df)
sns.set_style('darkgrid')

norm = plt.Normalize(df['average salary'].min(), df['average salary'].max())
sm = plt.cm.ScalarMappable(cmap="Greens", norm=norm)
sm.set_array([])
colors = sm.to_rgba(df['average salary'])
colors = [plt.cm.Greens(norm(value)) for value in df['average salary']]

# Создание графика
plt.figure(figsize=(10, 5))

# Построение столбчатой диаграммы с указанием hue и отключенной легендой
bar_plot = sns.barplot(x='Посада', y='average salary', data=df, palette=colors, legend=False)
if bar_plot.legend_:
    bar_plot.legend_.remove()

# Добавление значений над столбцами
for index, row in df.iterrows():
    plt.text(index, row['average salary'], round(row['average salary'], 2), color='black', ha="center")

# Настройка меток осей и заголовка
plt.xlabel("Посада", fontsize=12)
plt.ylabel("Середня заробітня плата", fontsize=12)
plt.title("Середня заробітна плата по ІТ спеціальностям", fontsize=15)

# Поворот меток оси X
plt.xticks(rotation=90)

# Отображение графика
plt.show()
```
